# Graph generation examples

## Import the graph generation functions

```python
import graph_gen
import networkx as nwx
```

## Complete graph generation

```python
complete_graph = graph_gen.gen_complete_graph(10)
nwx.draw(complete_graph, with_labels=True, font_weight='bold')
```

## Cyclic graph generation

```python
cyclic_graph = graph_gen.gen_cyclic_graph(10)
nwx.draw(cyclic_graph, with_labels=True, font_weight='bold')
```

## Random graph generation

```python
random_graph = graph_gen.gen_random_graph(10, 0.3)
nwx.draw(random_graph, with_labels=True, font_weight='bold')
```
